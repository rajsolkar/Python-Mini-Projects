```python
import random as rd
import time
operators = ["+","-","*"]
max_num = 10
min_num = 3
total_problems = 10
def generate_problem():
    left = rd.randint(min_num,max_num)
    right = rd.randint(min_num,max_num)
    operator = rd.choice(operators)
    expr = str(left) +" " +operator+ " "+  str(right)
    answer= eval(expr)
    return expr,answer
wrong = 0
input("Press Enter To Start")
print("----------------------------------")
start_time = time.time()
for i in range(total_problems):
    expr,answer = generate_problem()
    while True:
        user_ans = input("#Problem"+ str(i+1) +" " "Guess The Answer for  " + expr+":")
        
        if user_ans == str(answer):
            break
        
        wrong+=1
end_time = time.time()
total_time = end_time - start_time
print("Congrats You Finisehd in",total_time,"seconds")
print(f"You gave wrong answers {wrong} time(s).")
```
