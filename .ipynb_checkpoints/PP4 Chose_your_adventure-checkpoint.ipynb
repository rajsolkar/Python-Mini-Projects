<h1>The Lost City of Aetheria

```python
print("Welcome To The Lost City of Aetheria")
print("Where You are The Master Of Your Own Fate")
q1 = input("Ready To play The Lost City of Aetheria? Yes/No: ").lower()

if q1 == "yes":
    print("\nYou wake up in a strange forest, your memory foggy. A silver compass glows faintly in your hand, pointing toward a mountain in the distance.")
    print("1. Follow the compass toward the mountain.")
    print("2. Explore the forest for clues first.")
    q2 = int(input("Choose Option 1 or 2: "))

    if q2 == 1:
        print("\nAfter hours of climbing, you find an ancient stone gate carved with glowing runes.")
        print("1. Touch the glowing runes.")
        print("2. Look for another way around the gate.")
        q3 = int(input("Choose Option 1 or 2: "))

        if q3 == 1:
            print("\nYou are pulled into a portal and land in a forgotten city floating in the sky — Aetheria.")
            print("You see two paths: one leads to a golden tower, the other to a dark marketplace.")
            print("1. Enter the golden tower.")
            print("2. Explore the marketplace.")
            q4 = int(input("Choose Option 1 or 2: "))

            if q4 == 1:
                print("\nTower: You discover ancient archives and learn you’re destined to restore balance to Aetheria.")
            elif q4 == 2:
                print("\nMarketplace: You meet a mysterious merchant offering you a magical key in exchange for your compass.")
            else:
                print("\nInvalid Selection. Game Over.")

        elif q3 == 2:
            print("\nYou find a secret tunnel behind vines. It’s dark but leads deeper into the mountain.")
            print("1. Light a torch and enter.")
            print("2. Go back to the gate.")
            q6 = int(input("Choose Option 1 or 2: "))

            if q6 == 1:
                print("\nTorch path: You encounter glowing creatures that lead you to an underground lake full of crystals.")
            elif q6 == 2:
                print("\nGo back: The gate has disappeared — you’re now trapped in a time loop.")
            else:
                print("\nInvalid Selection. Game Over.")
        else:
            print("\nInvalid Selection. Game Over.")

    elif q2 == 2:
        print("\nYou find a strange traveler who claims to know you. He warns that the compass is cursed.")
        print("1. Trust him and hand over the compass.")
        print("2. Keep it and run toward the mountain.")
        q5 = int(input("Choose Option 1 or 2: "))

        if q5 == 1:
            print("\nThe traveler transforms into a guardian spirit and blesses you with a protective charm.")
        elif q5 == 2:
            print("\nThe compass burns your hand and creates a rift — pulling you into an alternate version of the forest where shadows whisper your name.")
        else:
            print("\nInvalid Selection. Game Over.")
    else:
        print("\nInvalid Selection. Game Over.")

elif q1 == "no":
    print("\nThank you for your time!")
else:
    print("\nInvalid Selection. Please enter Yes or No.")
```
