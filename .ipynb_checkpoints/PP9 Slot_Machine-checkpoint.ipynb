```python
## import random
MAX_LINES = 3
MAX_BET = 100
MIN_BET = 1
ROWS = 3
COLS = 3
symbols_count = {"A":2,"B":4,"C":6,"D":8}
symbols_values = {"A":5,"B":4,"C":3,"D":2}

def check_winnings(columns,lines,bet,values):
    winnings = 0
    winning_lines = []
    for line in range(lines):
        symbol = columns[0][line]
        for column in columns:
            symbol_to_check = column[line]
            if symbol != symbol_to_check:
                break
        else:
            winnings += values[symbol] * bet
            winning_lines.append(line + 1)

    return winnings, winning_lines
        
def spin_slot_machine(rows,cols,symbols):
    
        all_symbols = []
        for symbol,symbol_count in symbols.items():
            for _ in range(symbol_count):
                all_symbols.append(symbol)
        columns = []
        for _ in range(cols):
            column = []
            current_symbols = all_symbols[:]
            for _ in range(rows):
                value = random.choice(current_symbols)
                current_symbols.remove(value)
                column.append(value)
            columns.append(column)
        return columns
def print_slot_machine(columns):
    for row in range(len(columns[0])):
        for i, column in enumerate(columns):
            if i != (len(columns) - 1):
                print(column[row],end="  |  ")
            else:
                print(column[row])
        
def deposit():
    while True:
        ammount = input("Enter The Ammount You Want To Deposit:  ")
        if ammount.isdigit():
            ammount = int(ammount)
            if ammount > 0:
                break
            else:
                print("Enter Ammount Greater Than 0.")
        else:
            print("Please Enter A Numeric Value.")
    return ammount
def get_number_of_lines():
    while True:
        lines = input("Enter The Number Of Lines You Want To Bet On (1 - 3):  ")
    
        if lines.isdigit():
            lines = int(lines)
            if 1 <= lines <= MAX_LINES:
                break
            else:
                print("Enter Valid Number Of Line.")
        else:
            print("Please Enter A Numeric Value.")
    return lines
def get_bet():
    
    while True:
        bet = input("Enter The Ammount You Want To Bet on each line:  ")
        if bet.isdigit():
            bet = int(bet)
            if bet > 0:
                break
            else:
                print("Enter Ammount Greater Than 0.")
        else:
            print("Please Enter A Numeric Value.")
    return bet
def spin(balance):
    line = get_number_of_lines()
    while True:
        stake = get_bet()
        total_stake = stake * line
        if total_stake  > balance:
            print("Insufficient Balance.")
            
        else:
            break
    print(f"Your Total balance is ${balance}.\nYou Are Betting on {line} lines.\nYour betting ${stake} on each line.\nTherefore Your Total Betting Ammount is ${total_stake}")
    
    slots = spin_slot_machine(ROWS, COLS, symbols_count)
    print_slot_machine(slots)

    winnings, winning_lines = check_winnings(slots, line, stake, symbols_values)
    print(f"You won ${winnings} on lines {winning_lines}")
    return winnings - total_stake
def main():
    balance = deposit()
    while True:
        answer = input("Press Enter to Play or Q to exit").lower()
        print("\n-----------------------------")
        if answer == "q":
            
            break
        balance += spin(balance)
        if balance < 0:
            balance = 0
        print(f"Your Updated Balance is ${balance}")


main()

            
        
                
    
                    
```
