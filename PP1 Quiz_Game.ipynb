```python
user = input("Do you Want To Play A Quiz Game?  ")
if user.lower() == "yes":
    quit()
    

print("Every Correct answer gives you 10 points!" )
score = 0
answer = input("What is the first month of the year 1976?  ")
if answer.lower() == "january":
    print("Correct")
    score += 10
else:
    print("Incorrect")
answer = input("Which Country is the Coffee Pot Of the World?  ")
if answer.lower() == "brazil":
    print("Correct")
    score += 10
else:
    print("Incorrect")
answer = input("Which Country is the Largest By Land in the World?  ")
if answer.lower() == "russia":
    print("Correct")
    score += 10
else:
    print("Incorrect")      
answer = input("Which City is Most Popular in USA ?  ")
if answer.lower() == "new york":
    print("Correct")
    score += 10
else:
    print("Incorrect")
answer = input("Which Country is the smallest by land in the World?  ")
if answer.lower() == "vatican city":
    print("Correct")
    score += 10
else:
    print("Incorrect")
print("The End")
print(f"Your Total Points are : {score}")
print("Thanks For Playing.")

```
