<H1>Joins Answers:

```python
select c.customer_id,c.first_name , c.last_name,o.item from Customers c inner join Orders o on c.customer_id = o.customer_id;
select o.customer_id,o.order_id,c.country from Orders o left join Customers c on c.customer_id = o.customer_id;
select o.order_id,o.customer_id,o.item,o.amount,s.status from Orders o left join Shippings s
on o.customer_id = s.customer;
select distinct c.*,o.order_id,s.shipping_id from 
Customers c inner join Orders o
on c.customer_id = o.customer_id 
inner join Shippings s 
on c.customer_id = s.customer;
select c.* from Customers c left join Orders o on c.Customer_id = o.customer_id where o.order_id is null; 
```

<H1>Aggregations Answers: 

```python
select customer_id,sum(amount) as Total_Amount_Spent from Orders group by customer_id;

select avg(o.amount) as avg_amount,c.country from Orders o join Customers c on 
o.customer_id = c.customer_id where c.country = "UK";

select customer_id,count(order_id) as total_orders from Orders group by customer_id;

select customer_id,max(amount) as Max_order from Orders;

select sum(amount) as Total_Revenue from Orders;
```

```python
select * from Orders where amount > 500;

select country, avg(age) as avg_age from Customers group by country having avg(age) > 25 ;

select customer_id, count(order_id) as total_orders from Orders group by customer_id having 
count(order_id) > 1;

select * from Shippings where status = "Pending" ;

select c.*,sum(o.amount) as total_spending from Customers c inner join Orders o 
on c.customer_id = o.customer_id group by c.customer_id order by total_spending desc limit 2;
```

<h1>Joins Answer:

```python
select * , count(order_id) as total_orders from Orders group by customer_id;

select * from Shippings s left join Orders o on s.customer = o.customer_id where order_id is null;

select c.*,o.amount,s.status from Customers c 
inner join Orders o 
on c.customer_id = o.customer_id
inner join Shippings s
on c.customer_id = s.customer
where o.amount in (300,400);

select c.*,o.*,s.status from
Customers c 
left join Orders o on c.customer_id = o.customer_id
left join Shippings s on c.customer_id = s.customer;

select c.*,o.item from Customers c 
left join Orders o on c.customer_id = o.customer_id where c.country = "UAE";

```

<h1> Aggregations + Groups

```python
select customer_id,min(amount) as min_order_per_customer from Orders group by customer_id;

select country,avg(age) as avg_age from Customers group by country order by avg_age desc;



select c.*, sum(o.amount) as total_spending from Customers c inner join Orders o on c.customer_id = o.customer_id group by c.customer_id having 
sum(o.amount) < ( SELECT AVG(t.customer_total)
FROM ( SELECT SUM(amount) AS customer_total FROM Orders GROUP BY customer_id) AS t);
                 
select item,count(item) as total_orders from Orders group by item having total_orders > 1;
                 

select c.customer_id,c.first_name, sum(o.amount) as total_spending ,
round((sum(o.amount)*100.0) / (select sum(amount) from Orders),2) as percentage_contribution from Customers c inner join Orders o on c.customer_id = o.customer_id group  by c.customer_id;
```

<h1> Filtering with Aggregates

```python
select c.*,count(o.customer_id) as total_orders_by_customer
from Customers c inner join Orders o on 
c.customer_id = o.customer_id group by o.customer_id having total_orders_by_customer = 1;

select country,count(customer_id) as customers_from_country from Customers group by country having customers_from_country < 2 ;


select customer_id,max(amount) as max_amount,min(amount) as min_amount from Orders group by customer_id having max(amount) > (min(amount) * 2);

select c.* , s.status from Customers c inner join Shippings s on c.customer_id = s.customer
where s.status = "Pending";


SELECT c.*, COUNT(o.order_id) AS total_orders_by_customer
FROM Customers c 
INNER JOIN Orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id
HAVING COUNT(o.order_id) > (
    SELECT AVG(t.order_count)
    FROM (
        SELECT COUNT(order_id) AS order_count
        FROM Orders
        GROUP BY customer_id
    ) AS t
);

```

<h1> Subqueries / Nested Queries

```python

select *,avg(age) from Customers group by country  having age > avg(age);

select c.*,o.item from Customers c inner join Orders o on c.customer_id = o.customer_id
where o.item in  (select item from Orders where customer_id = 1);

 
 select c.* from Customers c where c.customer_id Not in
 (select o.customer_id from Orders o inner join Customers cu on o.customer_id = cu.customer_id where o.item in 
  ( select o2.item from Orders o2 inner join Customers c2 on o2.customer_id = c2.customer_id where c2.country  = "UK")
 );
 
select c.*,sum(o.amount) as total_spending from Customers c join Orders o on c.customer_id = o.customer_id 
group by c.customer_id having sum(o.amount) > (select max(amount) from Orders);

select c.*,s.shipping_id,sum(o.amount) from Customers c 
join Shippings s on c.customer_id = s.customer 
join Orders o on c.customer_id = o.customer_id 
group by c.customer_id having sum(o.amount) < 
(select avg(o2.amount) from Orders o2 inner join Customers c2 on
 c2.customer_id = o2.customer_id where c2.country = c.country);
```
