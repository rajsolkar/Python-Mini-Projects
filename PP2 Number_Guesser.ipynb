<H1>Final Code

```python
import random as rd

computer = rd.randint(1, 10)
answer = input("Want To Guess The Random Number Between 1 to 10? yes/no: ")
guesses = 0
if answer.lower() == "yes":
    while True:
        guesses += 1
        user = int(input("Enter Your Guess: "))
        if user == computer:
            print(f"Congrats, You Guessed It Right in just {guesses} guesses!!")
            print(f"The Random Number was: {computer}")
            break
        else:
            print("Sorry, You Guessed It Wrong.")
            ans2 = input("Want To Guess Again? yes/no: ")
            if ans2.lower() != "yes":
                print(f"The Random Number was: {computer}")
                print("Better luck next time!")
                break

elif answer.lower() == "no":
    print("Thank you for your time!")

else:
    print("Please enter yes or no.")
```
