<h1> The Pig Game

<h4>Pig Game Rules (Simple Version)

Two players take turns rolling a die.

On each turn:

The player can roll the die as many times as they like.

Each roll adds to their turn total.

But if they roll a 1, they lose all points for that turn, and it becomes the other player’s turn.

The player can choose to “Hold”, adding their turn total to their overall score and passing the turn.

First player to reach 100 points wins!

```python
import random

def roll():
    
    return random.randint(1, 6)


play_game = input("Ready to play the Pig Game? (Yes/No): ").lower()

if play_game == "yes":
    
    while True:
        players_input = input("Enter the number of players (2 - 4): ")
        if players_input.isdigit():
            players = int(players_input)
            if 2 <= players <= 4:
                break
            else:
                print("Must be between 2 - 4 players.")
        else:
            print("Invalid input, please enter a number.")

    max_score = 50
    player_scores = [0 for _ in range(players)]
    print("\nInitial scores:", player_scores)

    
    while max(player_scores) < max_score:
        for player_idx in range(players):
            print(f"\nPlayer {player_idx + 1}'s turn has started!")
            print(f"Current total score: {player_scores[player_idx]}")
            current_score = 0

            
            while True:
                should_roll = input("Do you want to roll the dice? (Y/N): ").lower()
                if should_roll != "y":
                    break

                value = roll()
                if value == 1:
                    print(f"You rolled a 1! Your turn score is reset to 0.")
                    current_score = 0
                    break
                else:
                    current_score += value
                    print(f"You rolled a {value}. \nCurrent turn score: {current_score}")

            player_scores[player_idx] += current_score
            print(f"Player {player_idx + 1}'s total score is now: {player_scores[player_idx]}")

            
            if player_scores[player_idx] >= max_score:
                print(f"\n Player {player_idx + 1} wins with a score of {player_scores[player_idx]}! ")
                exit()

elif play_game == "no":
    print("Thank you for your time!")
else:
    print("Not a valid selection.")
```
