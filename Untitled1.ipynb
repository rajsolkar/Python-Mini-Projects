```python
import math
def prime_numbers(n):
    prime_num = []
    for i in range(2,n+1):
        is_prime =True
        for j in range(2,int(math.sqrt(i))+1):
            if i%j == 0:
                is_prime = False
                break
        if is_prime:
            prime_num.append(i)
    return prime_num
prime_numbers(20)
    
```
