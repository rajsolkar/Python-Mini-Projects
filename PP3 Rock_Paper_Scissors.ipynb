```python
import random as rd
user = 0
computer = 0
game = ["rock","paper","scissors"]
comp_choice = rd.choice(game)
q1 = input("Ready To play Rock Paper Scissors?  Yes/No  ")
lower_q1 = q1.lower()
if lower_q1 == "yes":
    while True:
        inp1 = input("Enter Rock/Paper/Scissors Or Q to Exit")
        lower_inp1 = inp1.lower()
        
        if lower_inp1 == "q":
            print(f"Your Score: {user}")
            print(f"Computer Score: {computer}")
            
            break
        if lower_inp1 not in game:
            continue
        if lower_inp1 == "rock" and comp_choice == "scissors":
            print("You Won")
            print(f"Computer Chose: {comp_choice} ")
            
            
            user += 1
        elif lower_inp1 == "paper" and comp_choice == "rock":
            print("You Won")
            print(f"Computer Chose: {comp_choice} ")
            user += 1
        elif lower_inp1 == "scissors" and comp_choice == "paper":
            print("You Won")
            print(f"Computer Chose: {comp_choice} ")
            user =+ 1
        elif lower_inp1 == comp_choice:
            print("It's a Tie")
            print(f"Computer Chose: {comp_choice} ")

        else:
            print(f"Computer Chose: {comp_choice} ")
            print("You Lost")
            
            computer =+1
        
elif lower_q1 == "no":
    print("Thankyou For Your Time")
else:
    print("Please Enter Yes Or No")


```
